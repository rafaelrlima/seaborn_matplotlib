# Seaborn vs Matplotlib

```python
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Seaborn

#### Multigrid

Fonte: https://seaborn.pydata.org/tutorial/axis_grids.html

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip",  alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="sex", row="smoker", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip",  alpha=.7)
g.add_legend()
```

## Matplotlib (em construo)

#### Separando valores em 3D (em construo)
Fonte: https://stackabuse.com/matplotlib-scatterplot-tutorial-and-examples

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#%matplotlib notebook
%matplotlib inline

#se colocar opo inline no fica interativo
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150


# Add a 3D Scatterplot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
total_bill = tips['total_bill']
tip = tips['tip']

ax = fig.add_subplot(111, projection='3d')
ax.scatter(total_bill,tip, c=np.linalg.norm([total_bill,tip], axis=0))

plt.show()
```
